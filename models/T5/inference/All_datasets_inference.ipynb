```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from simplet5 import SimpleT5
import pickle
```

```python
# edit name of model file accordingly for different models
path_to_load_pickle_file="factuality_of_text_simplification/models/T5/training/T5_BERT_final.pkl"
Pkl_Filename = path_to_load_pickle_file
model = pickle.load(open(Pkl_Filename, 'rb'))
```

```python
import torch

model.device= torch.device("cpu")
```

```python
path_to_read_test_data="factuality_of_text_simplification/datasets/test_data/test.csv"
test_df=pd.read_csv(path_to_read_test_data,usecols=['source_text', 'simplified_text'])
```

```python
test_df['source_text'] = "simplify: " + test_df['source_text']
test_df
```

```python
# model.predict(df['source_text'])
import tqdm
model_output = []
for index, row in tqdm.tqdm(test_df.iterrows()):
    model_output.append(model.predict(row['source_text']))
```

```python
df_output = pd.DataFrame()
df_output["source_sentence"] = test_df['source_text']
df_output["gold_sentence"] = test_df['simplified_text']
df_output["pred"] =model_output 
```

```python
df_output
```

```python
temp_source_sentences= list(df_output['source_sentence'])
temp_source_sentences = [x.split() for x in temp_source_sentences]
temp_source_sentences = [(" ").join(x[1:]) for x in temp_source_sentences]
df_output['source_sentence'] = temp_source_sentences
```

```python
df_output
```

```python
pred_new = []
for idx, rows in df_output.iterrows():
    pred_new.append(" ".join(list(rows['pred'])))
pred_new
```

```python
df_output['pred_new'] = pred_new
```

```python
df_output
```

```python
df_output=df_output.drop(columns=['pred'])
df_output=df_output.rename(columns={"pred_new": "pred"})
```

```python
df_output
```

```python
# edit path and file names accordingly 
path_to_save_file="factuality_of_text_simplification/models/T5/inference/output_files/bert_score_model_inference.csv"
df_output.to_csv(path_to_save_file,index=False)
```
