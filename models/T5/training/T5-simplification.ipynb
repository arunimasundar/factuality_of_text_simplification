```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from simplet5 import SimpleT5
```

```python
path_to_load_data="factuality_of_text_simplification/datasets/final_training_data/bert_score_dataset.csv"
path_to_load_data="factuality_of_text_simplification/datasets/final_training_data/bart_score_dataset.csv"
path_to_load_data="factuality_of_text_simplification/datasets/final_training_data/dae_score_dataset.csv"
path_to_load_data="factuality_of_text_simplification/datasets/final_training_data/baseline_dataset.csv"
path_to_load_data="factuality_of_text_simplification/datasets/final_training_data/merged_dataset.csv"
df=pd.read_csv(path_to_load_data,usecols=['source_sentence', 'simplified_sentence'])
```

```python
df
```

```python
df = df.rename(columns={"simplified_sentence":"target_text", "source_sentence":"source_text"})
df = df[['source_text', 'target_text']]
```

```python
df
```

```python
df['source_text'] = "simplify: " + df['source_text']
df
```

```python
train_df, val_df = train_test_split(df, test_size=0.2)
train_df.shape, val_df.shape
```

```python
model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_df,
            eval_df=val_df, 
            source_max_token_len=128, 
            target_max_token_len=128, 
            batch_size=8, max_epochs=5, use_gpu=True)
```

```python
# edit path accordingly for T5 models on different datasets
path_to_save_pickle_file="factuality_of_text_simplification/models/T5/training/T5_BERT_final.pkl"
import pickle
Pkl_Filename = path_to_save_pickle_file 

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(model, file)
```
